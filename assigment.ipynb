**Q1 . Prime Number Checker with Command-Line Arguments**

* Write a Python script that checks if a given number is prime. The script should accept a number as a command-line argument and print whether the number is prime or not.


    

```python
import sys

def is_prime(n):
    """Check if a number is prime."""
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

if __name__ == "__main__":
    if len(sys.argv) != 2:
        print("Usage: python script.py <number>")
    else:       
        try:
            num = int(sys.argv[1])
            if is_prime(num):
                print(f"{num} is a prime number.")
            else:
                print(f"{num} is not a prime number.")
        except ValueError:
            print("Please enter a valid integer.")
```

**Q3.List Iteration with Skipping, Stopping, and Exception Handling**

Problem Description

* Write a Python program that iterates through a list and:

* Skips negative numbers

* Stops execution when zero is encountered

* Handles non-integer values using exception handling


```python
def process_list(lst):
    for item in lst:
        try:
            num = int(item)
            if num < 0:
                print(f"Skipping negative number: {num}")
                continue
            elif num == 0:
                print("Encountered zero. Stopping execution.")
                break
            else:
                print(f"Processing positive number: {num}")
        except ValueError:
            print(f"Non-integer value encountered: {item}")


my_list = [10, -5, 3, 0, 8, "hello", 7.5]
process_list(my_list)
```

**Q4.List Comprehension for Squares Divisible by 2 and 5**

***Problem Description***

Write a Python program using a single list comprehension to generate a list of squares of numbers from 1 to 20 that are divisible by both 2 and 5.

Concept Explanation

* A list comprehension is a concise way to create lists in Python.

* A number divisible by both 2 and 5 must be divisible by 10.

* The program checks this condition and then stores the square of such numbers.

```python
squares = [x**2 for x in range(1, 21) if x % 2 == 0 and x % 5 == 0]
print(squares)
```

**Q5. Find Minimum and Maximum in a List**

***Problem Statement***

* Write a Python program to find the smallest and largest elements in a list using a user-defined function, without using built-in functions like min() and max().

```python
def find_min_max(lst):
    if not lst:
        return None, None
    smallest = largest = lst[0]
    for num in lst:
        if num < smallest:
            smallest = num
        if num > largest:
            largest = num
    return smallest, largest
numbers = [34, 7, 23, 32, 5, 62]
smallest, largest = find_min_max(numbers)
print(f"Smallest: {smallest}, Largest: {largest}")
```
